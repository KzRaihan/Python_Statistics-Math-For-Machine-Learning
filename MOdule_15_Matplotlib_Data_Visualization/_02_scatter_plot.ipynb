### 1. Introduction of Scatter Plot and Why Usage Scatter plot in ML.

- Scatter plot find the correlation between features.

- Correlation means two or more features relation.
    - 1. Positive Correlation
    - 2. Negative Correlation
    - 3. No Correlation
        
- Why Usage Scatter plot in ML
    - 1. To select the Model Algorithm.
    - 2. To Select the features for training the model.
    - 3. To Detection the outliers.

- if  features Correlation is Positive, negative then we count those features to train the model.


- if features has no Correlation then we can't count this features to train the model

```python
# import libraries
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
# load csv file
df = pd.read_csv('student_IQdata.csv')

df.head(8)
```

- We try to build a model that Predict the IQ_Score as per the input feature

- so, first we see which input features are Correlation between IQ_Score.

#### First try to see the realation between feature Shoes_Size and IQ_Score

```python
# if we try to 2-d ploat 
plt.plot(df['IQ_Score'], df['Shoe_Size'])


plt.show()
```

```python
# Use scatter plot to find correlation

plt.xlabel("IQ Score")
plt.ylabel("Shoe Size")
plt.title("Relation between IQ Score and Shoe Size")

plt.scatter(df['IQ_Score'], df['Shoe_Size'])


# plt.legend()
plt.show()
```

- So, According to see the plot there is no Relation between IQ Score and Shoe Size

```python
df.head(2)
```

#### 2. Now Try to see the Relation between IQ Score and Study_Hour features

```python
plt.xlabel("IQ_Score")
plt.ylabel("Study_Hour")
plt.title("Relation between IQ Score and Study Hours")


plt.scatter(df['IQ_Score'], df['Study_Hour'], color = 'r')


plt.show()
```

- According to see the plot there is positive Correlation between IQ Score and Study Hours

- why?
    - here When Study hours are increases then IQ score are increases that way it's positive correlation.

#### 4. Now we try to see the Relation between IQ_Score and Chilling_Hours features

```python
plt.xlabel("IQ_Score")
plt.ylabel("Chilling_Hours")

plt.title("Relation between IQ Score and Chilling Hours")


plt.scatter(df['IQ_Score'], df['Chilling_Hours'],color='g')

plt.grid()
plt.show()
```

- According to see the plot there is a negative Correlation.

- Why?
    - Because when Chilling Hours are increases then IQ Score in decreases.

### Overall:

- To Build a model we ignore those features that have no correlation.

- only count those featues that have a correlation (positive or negative).
