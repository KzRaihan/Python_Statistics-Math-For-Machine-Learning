### 1. What and Why use Bar chart

- What?
    - A bar chart is used to represent categorical data using rectangular bars.


- Why?
    - To compare data between categories.



```python
# Syntax of Bar chart

'''  
- plt.bar(x_axix, y_axix, color=None, width = None)

- x_axis => Categories
- y_axis => Frequency (no. of values)

'''
# x_axis(categories) -> first group the target column
#     - then target column index.size
# y_axis
#      - target column value 
```

### 2. Example of Bar chart

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python

data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df.head(3)
```

```python
# bar chart as gender wise

group_gender = df.groupby('Gender').size()


print(group_gender.index) # this is x_axis
group_gender.values  # this is y_axis
```

```python
plt.xlabel("Gender of Student")
plt.ylabel("count of Student")
plt.title("Bar chart of Gender among students")

plt.bar(group_gender.index, group_gender.values)


plt.show()
```

```python
# customize the color and others
plt.xlabel("Gender of Students")
plt.ylabel("Frequency of Students")
plt.title("Bar chart of Gender among students")

plt.bar(group_gender.index, group_gender.values,color=['blue', 'green'], edgecolor = 'black')

plt.show()
```

#### Bar chart form real df

```python
# load df

df = pd.read_csv("student_data.csv")

df.head(1)
```

```python
# bar chart based on CompletionStatus column

# first group this CompletionStatus column

df_group = df.groupby('CompletionStatus').size()

df_group.index

df_group.values
```

```python
# Create bar chart on CompletionStatus column

plt.xlabel("Completion Status")
plt.ylabel("Count of students")
plt.title("Bar chart of Course Completion")

plt.bar(df_group.index,df_group.values, color =['g','b','r'],edgecolor='black')

plt.show()
```

```python
# bar chart of location wise

# first group the location column
grop_loc = df.groupby('Location').size()

# bar chart

plt.xlabel("Location")
plt.ylabel("Student Frequency")
plt.title("Bar chart of Location based on Student")

plt.bar(grop_loc.index, grop_loc)

plt.show()
```

```python
# bar chart based on Instructor
# bar chart of Instructor wise

# first group the Instructor column
grop_loc = df.groupby('Instructor').size()

# bar chart

plt.xlabel("Instructor")
plt.ylabel("Student Frequency")
plt.title("Bar chart of Instructor based on Student")

plt.bar(grop_loc.index, grop_loc, color=['r','b','g'], edgecolor = 'black')

plt.grid()
plt.show()
```
