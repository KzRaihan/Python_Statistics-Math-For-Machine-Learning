### 1. What is Histogram and why it's Use?

- What:
    - A histogram is a graphical Representation of the distribution of a dataset.


- Why usage?
    - To understand the data distribution over a factor.
    - To Analysis shape (Symmetric,left/right skewed) of datasets.
    - To Detect outliers.


- Why need in ML?
    - when we build a model if datasets left,right skewed then our model biased prediction.

    - when datasets are Symmetric distribution then our model predictionis less biased.

    - Also if dataset contain more outliers then model prediction is biased.


    - So, To see the datasets to analysis and make decision, like this datasets are perfect or not to train the model.


```python
# Syntax of Histogram
'''   
- plt.hist(x, bins = None, color = 'None', edgecolor = None)

here,
    - x = Numeric data
    - bins = Numbers of division (default = 10)

'''
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_IQdata.csv')

print(df.head(2))

df.tail(2)
```

```python
# histogram based on IQ Sore

plt.xlabel("IQ Score")
plt.ylabel("Frequency of Student")

plt.title("Student IQ Distribution")

plt.hist(df['IQ_Score'], color='g', edgecolor = 'black')


# plt.grid()
plt.show()
```

- The above histogram is Right skewed Distribution

```python
# Histogram based on Study Hour

plt.xlabel("Study Hours")
plt.ylabel("Frequency of Student")
plt.title("Student Study Hour Distribution")

plt.hist(df['Study_Hour'],color='b', edgecolor='red')


plt.grid()
plt.show()
```

- According to see this graph this is symmetrical distribution.

```python
df.head(2)
```

```python
# Histogram based on Chilling hours

plt.xlabel("Shoe size")
plt.ylabel("Frequency of Student")
plt.title("Shoe size Distribution")

plt.hist(df['Shoe_Size'], color = 'r', edgecolor='black')


# plt.grid()

plt.show()
```

- According the see the graph this is right skewed distribution

### Customize the bins

- Number of division (default = 10)

```python
df.head(2)
```

```python
# customize the bins

plt.xlabel("Chilling Hours")
plt.ylabel("Frequency of Students")
plt.title("Student Chilling Hour Distribution")

plt.hist(df['Chilling_Hours'],bins=35)


plt.show()
```

- Here, number of bins (x-axis) is 35.

```python
# Customize the bins

plt.xlabel("Study Hours")
plt.ylabel("Frequency of Student")
plt.title("Study Hours Distribution")


plt.hist(df['Study_Hour'],bins=[1,2,3,4,5,6,7], color='#012C22', edgecolor ='#9810FA')

plt.grid()
# plt.grid(axis = 'y', linestyle='--')
plt.show()
```

- Explaination of parameters
    - bins =[1,2,3,4,5,6,7] bin edges meaning
    - (1-2) => 1st bar
    - (2-3) => 2nd bar and so on.
