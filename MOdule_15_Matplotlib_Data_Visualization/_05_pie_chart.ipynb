## Topic: Pie chart in pandas

- what?

    - A pie chart is a circular chart divided into slices that represent percentages of a whole dataset.

    - each slice show hom much one category contributes to the total.


- Why?
    - Percentage of expenses in a monthly budget

```python
# Syntax of pie chart
'''  
plt.pie(group_column, label = group_column_index, autopact = None, explode = (None), shadow = True)

'''
# here
# group_column => which column wise to see pie chart

# label => for each slice define the name

# autopact=> Display percentage
#    - autopact = '%1.1ff%%

# expolde => Seperate slice from the center
#  explode = (0.1,0,0) [among three group the first is seperate]


# shadow => adds a 3-d shadow effect
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_data.csv")

print(df.head(3))
df.tail(3)
```

```python
# apply pie chart for CompletionStatus column

# first group this column
group_cs = df.groupby('CompletionStatus').size()

# use pie chart
plt.title("Pie chart of Course Completion")

plt.pie(group_cs)

plt.show()
```

```python
# now level of each slices

# use pie chart
plt.title("Pie chart of Course Completion")

plt.pie(group_cs,labels=group_cs.index)

plt.show()
```

```python
# now lebels, pecentages of each slices
# use pie chart
plt.title("Pie chart of Course Completion")

plt.pie(group_cs,labels=group_cs.index, autopct='%1.1f%%')

plt.show()
```

```python
# now lebels, pecentages, separates one slice(explode) of each slices

plt.title("Pie chart of Course Completion")

plt.pie(group_cs,labels=group_cs.index, autopct='%1.1f%%', explode = (0.1,0,0)) # here three labels

plt.show()
```

```python
# now lebels, pecentages, separates one slice(explode) and shadow of each slices

plt.title("Pie chart of Course Completion")

plt.pie(group_cs,labels=group_cs.index, autopct='%1.1f%%', explode = (0.1,0,0),shadow = True) 

plt.show()
```

```python
# apply pie chart on Instructor column

# 1st perform -> group
group_ins = df.groupby('Instructor').size()

# customize the parameters
   # labels
idx = group_ins.index

    # each label color (here three label)
cols = ['gold', 'lime', 'salmon']

     # 2nd, 3rd slice separate(explode)
sep = (0,0.1,0.1)

plt.title("Pie chart of Instructor based on Student")

plt.pie(group_ins,labels=idx,autopct= '%1.1f%%', colors = cols,shadow = True, explode = sep)


plt.show()
```
