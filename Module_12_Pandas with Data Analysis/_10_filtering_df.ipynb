### 1. Intro of Filtering

- Extract Specific rows and subsets that meet certain logical conditions.

```python
# Filter rows where Data Structure Marks is greater then 90

df[df["Data Structure Marks"] > 90]
```

```python
'''  
- we are follow loc[] function

- loc[] is a label-based data selection method in pandas.

- Syntax: 
    - Df.loc[rows_condition, columns_selection]


'''
```

```python
# filter rows using loc[]

df.loc[df["CompletionStatus"] == "In Progress"]
```

```python
''' 
- How to work df.loc[df["CompletionStatus"] == "In Progress"]

Steps:
    - df.loc means Access all elements inside the  df 

    - df.loc[df["CompletionStatus"] means inside the all df to filter only CompletionStatus column (or only go seleted column)

    - df.loc[df["CompletionStatus"] == "In Progress"] means To check CompletionStatus column to "In Progress" string

    - if condition is meet then show or return the output otherwise not reture .


'''
```

### 2. Apply Conditional filtering

```python
df.head(2)
```

```python
# filter rows where Python Marks is less than or equal to 80 and show only StudentID and Location

df.loc[df["Python Marks"] <= 80, ["StudentID", "Location"]]
```

```python
# Filter multiple conditions using And(&) [both condition must be satisfy]

 # Q.n: find those students who status is 'Completed' the Course and 'Data Structure Marks' is greater than 90


compled_g_90 = df.loc[(df["CompletionStatus"] == "Completed") & (df["Data Structure Marks"] >= 90)]

compled_g_90
```

### 3. Filtering using logical operation

```python
# if the condition is satisfty then only show specific columns

# Q.n: find those students who Data Structure Marks is greate than or equal 90 and Python Marks is greater than 80 then only show StudentID,Instructor, Location , Python Marks, Data Structure Marks columns.

students = df.loc[(df["Data Structure Marks"] >= 90) & (df["Python Marks"] >= 80), ["StudentID", "Instructor", "Location", "Python Marks", "Data Structure Marks"]]

students
```

```python
# Q.n: find those students who Data Structure Marks is greate than and equal 90 and Python Marks is greater than 80 and Algorithm Marks is greater than 85  then only show StudentID,Instructor, Location , Python Marks, Data Structure Marks columns.


st_ds_py = df.loc[(
    # 1st condition
    (df["Data Structure Marks"] >= 90) &
    # 2nd condition
    (df["Python Marks"] >= 80) &
    # 3rd condition
    (df["Algorithm Marks"] >= 85) 

)]

st_ds_py

```

```python
# or operation (|)
st_ds_py = df.loc[
    (df["Data Structure Marks"] <= 70) |

    (df["Python Marks"] <= 80),

    ["StudentID","FullName"]]

st_ds_py
```

```python
data = {
    'Student': ['Rafi', 'Tina', 'Sami', 'Lina', 'Omar', 'Sara'],
    'Age': [18, 20, 19, 22, 21, 20],
    'Subject': ['Math', 'English', 'Physics', 'Math', 'Chemistry', 'Physics'],
    'Score': [85, 90, 70, 95, 88, 92]
}

df = pd.DataFrame(data)

df
```

```python
# 1. Filter students who scored above 85.

score_85 = df.loc[df["Score"] > 85]

score_85
```

```python
# 2. Filter students whose Subject is Math or Physics.
math_phy = df.loc[
    (df["Subject"] == "Math") |

    (df["Subject"] == "Physics")

    ]
math_phy
```

#### Negation in .loc[] Filtering

- inverts of .loc[] 
    - if condition = True return False
    - if condition = False return True
    - so, (~) negates or reverses that condition

- Syntax :
    - df.loc[~(condition)]

```python
df
```

```python
# filter those Student who Score less than 90 using nagation

ng_90 = df.loc[~(df["Score"] >= 90)]

ng_90
```

```python
# filter those Student who age greater than 20 using nagation

greater_90 = df.loc[~(df["Age"] <= 20)]

greater_90
```
