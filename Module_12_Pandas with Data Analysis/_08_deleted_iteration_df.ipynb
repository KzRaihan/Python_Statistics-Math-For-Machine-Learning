### 1. Deleted Single and Multiple Rows , columns


#### Delete Single Column

- Syntax of drop()

    - df.drop("coln_name", axis = 1, inplace = False)

    - axis = 1 menans column
    - inplace = False => return new copy df

```python
df.head(1)
```

```python
# df.drop() for single column

single_col_drop = df.drop("EnrollmentDate", axis = 1)

single_col_drop.head(3)
```

```python
# df.drop() for single column with inplace

single_col_drop = df.drop("StudentID", axis = 1)

single_col_drop.head(3)
```

#### Delete Multiple Column

- Syntax of drop()

    - df.drop(["coln1_name", "coln2"], axis = 1, inplace = True)

    - axis = 1 menans column.

    - inplace = True => return and updates DataFrame directly.

```python
single_col_drop.head(3)
```

```python
# Deleted multiple columns 

single_col_drop = single_col_drop.drop(["Instructor", "EnrollmentDate"],axis=1,inplace = True)

# IMPORTANT NOTE -> Run/Execute df.drop only one times 
# if we run df.drop multiple times then it's will show error.(in this case we try to deleted columns that are already deleted)
```

```python
single_col_drop.head(3)
```

```python
single_col_drop.drop("FullName", axis = 0,inplace = True)

# here The FullName rows is not exist 
```

### 2. Assign and access single and multiple values Using df.loc

- Syntax of df.loc[]

    - df.loc[row_level, col_level] = new_value

```python
df.head(2)
```

```python
# assign a single vale or update a single value in specific cell.

df.loc[1, "FullName"] = "KzRaihan"

df.head(3)


# (row,col) = value
# (1, FullName) =  KzRaihan

# cell(1,FullName) is change by value of  KzRaihan
```

```python
df.head(9)
```

```python
# access and update multiple cells values or rows values

df.loc[[0,2,4], ["FullName", "Algorithm Marks","CompletionStatus"]] = "Raihan", 96.50, "Completed"

df.head(8)

'''
for one element:
--------------------
# (0,FullName) = Raihan
# (0,Algorithm Marks) = 96.50
# (0,CompletionStatus) = Completed

for 2nd element:
--------------------
# (2,FullName) = Raihan
# (2,Algorithm Marks) = 96.50
# (2,CompletionStatus) = Completed

for 3rd element:
--------------------
# (4,FullName) = Raihan
# (4,Algorithm Marks) = 96.50
# (4,CompletionStatus) = Completed


'''
```

```python
# assign multiple row values

df.loc[[4,6], "Data Structure Marks"] = [90.45,90.90]

df.head(8)

'''
(4,Data Structure Marks) = 90.45
(6,Data Structure Marks) = 90.90

'''

```

```python
df.loc[[1], "Data Structure Marks"] = [2.45,9.99]

df

'''  
(1,Data Structure Marks) = 2.45
(1,Data Structure Marks) = 2.45 (x) show ValueError



'''
```

```python
df.head(2)
```

```python
# Entire row replace of new data

df.loc[1] = ["AI/ML006","KzRaihan", 98.88, 97.69,99.94,"Completed", "2025-08-24", "SK Rahim", "Dhaka"]

df.head()

# NOTE -> Assign all values of the columns otherwise ValueError
```

```python
# assing new column

df.loc["Performance"] = ["AL/ML_001", "Ra", 90.70, 95.50, 96.50, "In_Processing", "2025-08-05", "KzRaihan", "Dhaka"]

df

# Add a new row at the end
# here Performance = row idx
# must be the equal value of the columns
```

### 3. Iteration Over DataFrame

```python
df.head(3)
```

```python
# 1. Iterative over idx and series [ df.iterrows() ]

for idx, row in df.iterrows():
    print(f"{idx} : {row}")

# NOTE -> iteration over idx and series
```

```python
df.head(2)
```

```python
# Iteration over itertuple()

for i in df.itertuples():
    print(i)

```

```python
for i in df.itertuples(index = False):
    print(i)
```
