#### Access Multiple row using a fancy idx

- syntax:
    - df.loc[[r1, r2, rn], [c1, c2, cn]] 

- Note: idx start with zero

```python
df.loc[[2,4,1]] # 2, 4, 1 => idx's of rows 
```

#### Using slicing or range

- Synatax:
    - df.loc[start_row_idx : end_row_idx , start_col_id : end_col_idx]

    - default value:
        - start_row_idx = 0
        - end_row_idx = len(col)  <-- this is include

        - start_col_idx = o
        - end_col_idx = len(row)  <---- this is include

- NOTE for loc: end idx is include

```python
# Access Multiple data in a df

df.loc[:, : ]

# work as default values
```

```python
# customize the default values

df.loc[10:19, :]

# start row idx = 10 and end row idx = 19 (include)

# start column idx = 0 and end coln idx = len(coln) [include]
```

```python
df.loc[0:2]  # here col_idx = count as default values
```
