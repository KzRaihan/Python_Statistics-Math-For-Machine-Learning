### 1. DataFram in Pandas

- DataFram is Tabular (rows, columns) Structure data.

- Similar to 2-d Array

- collections of series

- syntax of datafram (df):

    - df = pd.read_csv("file_path")

    - pd => class of pandas

    - read_csv => read only csv file

    - file_path => csv file path in our machine

    - df => store file in Datafram stucture(rows, columns)

```python
# Basic example of DataFram

# read/load csv file

df = pd.read_csv("student_data (1).csv")

print("Type: ", type(df)) # < DataFram > 
df
```

### 2. Series in Pandas

- A pandas series is like a column or row in a table.

```python
# Example of serice

series = df["FullName"]

print("Type:", type(series)) # < 'Series' >

series


# in DataFram single one column is called series
```
