#### df.head(n)

- Return first n rows (Default 5)

- To quickly previews first few rows of a datafram.

- To confirm data loaded correctly

```python
df.head()
```

```python
df.head(3)
```

#### df.tail()

- To see the last few rows in a datafram.

- latest dates or transactions to verify if data loaded till the end.

```python
# df.tail(n=5) -> return last n values. [default (5)]

df.tail()
```

```python
# df.tail(3) -> return last 3 values

df.tail(3)
```

```python
# df.tail(8) -> return last 8 values

df.tail(8)

# df.tail(n) -> return n last values
```

#### df.sample(n=1)

- return randomly n values in a datafram (default = 1)

- It's help to explore data with baise.

```python
# df.sample()

df.sample()


# default -> 1 (show ramdomly 1 simple)
```

```python
# Example of df.sample(n)
# return random n rows in a datafram

df.sample(4)
```
