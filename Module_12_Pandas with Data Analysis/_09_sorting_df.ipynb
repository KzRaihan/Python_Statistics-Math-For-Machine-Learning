```python
df.head(3)
```

### 1. Sort Dataframe by Column Values

- syntax: 
    - df.sort_values(["col1", "col2"], ascending = True, inplace = False)

```python
# Sort by a single column with acending order
copy_df = df.sort_values("Python Marks", ascending= True)
copy_df.head(10)
```

```python
# Sort by multiple column with decending order

copy_df.sort_values(["Algorithm Marks","Data Structure Marks" ], ascending= False, inplace= True)

copy_df.head(10)
```

```python
df # there is not change
```

```python
## Sort by multiple columns with different orderings

# 'Data Structure Marks' =  decending order
# 'Python Marks'=  ascending order 	

new_df = df.sort_values(["Data Structure Marks", "Python Marks"], ascending=[False, True])

new_df.head(10)


# another format: ascending = [0, 1] 
# 0 means False (Decending order)
# 1 means True (ascending order)
```

### 2. Sort DataFrame by index (row lebels)

```python
df.head(4)
```

```python
# sort index in ascending order

df.sort_index(ascending=1)
```

```python
# sort rows by index in descending order

df.sort_index(ascending=0) # 0 => False => decending order

```

```python
# sort column names alphabetically

df.sort_index(axis=1).head(3)
```

```python
# sort column in reverse (alphabetically reverse) order

df.sort_index(axis=1, ascending = False).head(3)
```

```python
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Tablet'],
    'Price': [1200, 25, 45, 300, 250],
    'Stock': [10, 200, 150, 30, 60]
}
df = pd.DataFrame(data)

df
```

```python
#  1. Sort the DataFrame by Price (ascending).

df.sort_values("Price", ascending=1)
```

```python
# 2. Sort by Stock (descending).

df.sort_values("Stock", ascending=0).head(3)
```

```python
# 3. Sort by Product name alphabetically.

df.sort_values("Product", ascending=1)
```

```python
df # here no change
```

```python
# 4. Sort by Stock first (decending), then Price (ascending).

df.sort_values(["Stock","Price"], ascending=[0,1])
```
