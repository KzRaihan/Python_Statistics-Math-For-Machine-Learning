### 1. What/Where/Why  Scatter plot

- What?
    - A scatter plot used to visualize the relationship between two continuous variables.

    - Each data point is shown as a dot on the graph.


- Where?
    - Business, education, Healthcare, agriculture, ML and so on.


- Why?
    - Relationship detection.

    - Pattern Identification 

    - Outlier Detection.

    - Correlation Check

### 2. axes label Scatter plot using Seaborn

```python
# import libraries

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

```

```python
# load file
df = pd.read_csv("sns_data.csv")
df.head(3)
```

```python
# apply axas lebel scatter plot
# scatter plot use to find the correlation between numerical features (study_hours column and test_score)

sns.scatterplot(data = df, x = 'study_hours', y='test_score')

plt.show()
```

```python
df.head(1)
```

```python
# customize the axes label scatter plot parameters

sns.scatterplot(data = df, x = 'study_hours', y = 'test_score', hue = 'gender')


plt.show()

# hue => how many types of people (category)
```

```python
df.head(1)
```

```python
# customize the axes label parameter for scattter plot

sns.scatterplot(data = df, x='study_hours', y='test_score',hue = 'gender', style = 'class_level', size = 'Tshirt_size')
```

- Here,
    - data => DataFrame (work on this data).
    
    - x => respeted column
    
    - y => changes column
    
    - hue => Total category or group (types of people)
    
    - style => dot style according to parameter (here class label)

    - size => Quantity (Number of People)

```python
# scatter plot based on figure label

sns.relplot(
    kind = 'scatter', data = df,
    x = 'attendance_rate', y = 'test_score',
    hue = 'gender'
)

plt.show()

# positive correlation between attendence_rate , test_sore
```

```python
df.head(1)
```

```python
# customize the fingure label scatter plot parameters


sns.relplot(
    kind = 'scatter', data = df,
    x = 'study_hours', y = 'test_score',
    hue = 'gender',
    style = 'gender',
    size = 'Tshirt_size'
)

plt.show()
```

- Explain:
    - hue => group of person (here, gender)
    - style => dot style in graph
    - size => total number of person (T_shirt)

                - for T_shirt size (quantity)

                - 10 student T_Shirt size => tiny 

                - 20 student T_shirt size => small 

                - 30 student T_shirt size  => medium 

                - 40 student T_shirt size => large

### 4. Seaborn in-Bulid Dataset

```python
# load real in build dataset from online

df_tips = sns.load_dataset("tips")

df_tips.head()
```

```python
# find the correlationship between numerical columns
# - total_bill column and tip column


# 1. axes label 

sns.scatterplot(
    data = df_tips,
    x = 'total_bill', y = 'tip',
    hue = 'sex',
    style = 'time'
)
```

```python
df_tips.head(10)
```

```python
# 2. figure label for In-build dataset

sns.relplot(
    kind = 'scatter', data = df_tips, 
    x = 'total_bill', y = 'tip',
    hue = 'sex',
    style = 'time',
    size = 'size'
)

plt.show()
# size => total number of people
```
