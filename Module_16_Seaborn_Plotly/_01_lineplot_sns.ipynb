### 1. Where/Why use lineplot?

- A line plot in seaborn is used to visualize continuous data trends.


- Where it's used:
    - For time delta data.
    - likes (Business, Education, Helthcare, Weather, Finance)


- Why use line plot
    - lines to show how one variable changess with respect to another.

    - shows trends clearly

```python
# Syntax of lineplot using seaborn

'''  
- 1. Axes Label:

    sns.lineplot(data = df, x_axis = 'column1', y_axis = 'column2', errorbar = None)


'''

# here 
# df = DataFrame
# x_axis => respected value
# y_axis => changes value.
```

### 2. axes lebel plot using seaborn

```python
# import module

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# load file
df = pd.read_csv('sns_data.csv')

df.head()
```

```python
# 1. Axes line plot using seaborn

sns.lineplot(data = df, x = 'week', y = 'attendance_rate')

plt.show()
```

```python
# customize the parameters on axes seaborn lineplot

sns.lineplot(data = df, x = 'week', y = 'attendance_rate', errorbar = None)
```

```python
# customize the parameters on axes seaborn lineplot

sns.lineplot(data = df, x='week', y='attendance_rate', errorbar = None, hue = 'gender')


# show multiple line graph according to the hue or gender columns
```

### 3. figure label plot using seaborn

```python
# syntax
'''  
sns.relplot(

        kine = 'line', data = df

        x = 'column_of_respected', y='column_of_change',

        hue = 'column_to_show_multiple_line',

        errorbar = None       
        
)

'''
# kine => which type of plot 
# errorbar => no need to show shadow
```

```python
df.head(2)
```

```python
# 2. example of figure label lineplot in seaborn

sns.relplot(kind = 'line',data = df, x ='week',y='attendance_rate', hue = 'gender', errorbar = None )
```

```python
''' 
- According to the above graph we say
    - Male attendance is higger than female

    - day     |    gender     | attendance_rate
    --------------------------------------------------

    - 1       |     male      |  91%

    - 1       |     Female    |  85%


    .......................................

    - 4       |     male      |  90%

    - 4       |     Female    |  85%

    
    
    - 6       |     male      |  75%

    - 6       |     Female    |  62.5%

'''
```

```python
# customize the parameters 

sns.relplot(kind = 'line', data = df, x='week', y ='attendance_rate', hue = 'class_level', errorbar = None)
```

### 4. Different between axes and figure label line plot in seaborn.

- Axes label line plot:
    - Inside the figure label the single part or portain is called Axes Label.


- Figure Label line plot:
    
    - The entire box is called figure
    - One figure can have multiple axes
