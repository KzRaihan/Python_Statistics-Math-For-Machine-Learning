### 1. DateTime

- DataTime represent data and time values in a structured Format.

- Extracting year, month, day, hour etc.

- syntax:
    - pd.to_datetime(data,format = None)

    - data => A column, or Series of time string format(object format)

    - format => yera-month-day

```python
import pandas as pd
```

```python
# load file
df = pd.read_csv('student_completed_data.csv')

df.head(3)
```

```python
# To check the Data type of column
df.info()
```

```python
# first convert data type object format to datatime64 data format 

df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])

df.head()
```

```python
df.info()

# here, EnrollmentDate data type is convert to datetime64 format
```

```python
# now, Extract Date Components 
# Year

df['Year'] = df['EnrollmentDate'].dt.year

df.head()
```

```python
# Extract Date Component 
# month

df['Month'] = df['EnrollmentDate'].dt.month

df.head(3)
```

```python
# Extract Date Component
# day

df['Day'] = df['EnrollmentDate'].dt.day

df.head(4)
```

### TimeDelta in Pandas

- Represent the difference between two datatime values(duration or time gap)

- Syntax:
    - first convert object data type to datetime64 data type
    - pd.to_timedate(data)

    - 2nd use TimeDelta:
        - pd.to_timedelta(arg, unit = 'd')

```python
# first conver object data type to datetime64

df['FinishedDate'] = pd.to_datetime(df['FinishedDate'])

df.head()
```

```python
df['FinishedDate'].info()

# Type -> datetime64
```

```python
df['Time Taken for Finish'] = df['FinishedDate'] - df['EnrollmentDate']

df.head(5)
```
