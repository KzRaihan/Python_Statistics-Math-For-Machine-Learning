```python
df = pd.read_csv("student_data.csv")

df.head()
```

### 1. Basic Statistical Functions

- Syntax:
    - df[['col1', 'col2']].sum(axis = 1)

    - axis = 1 => column wise 
    - axis = 0 => row wise (By default)

```python
# fill NaN value with zero
df.fillna(0)
```

```python
# Sum of Python Marks 

df['Python Marks'].sum()	

# sum of Python Marks column
```

```python
# min value in Python Marks 

df['Python Marks'].min()
```

```python
# max value in Python Marks

df['Python Marks'].max()
```

```python
# mean value in Python Marks
df['Python Marks'].mean()
```

```python
# median value in Python Marks

df['Python Marks'].median()
```

```python
# std value in Python Marks

df['Python Marks'].std()
```

- Descriptive Statistics Summary

```python
df.describe()
```

### Sum accross Multiple columns

```python
# Multiple Columns sum() of total

df[['Data Structure Marks', 'Algorithm Marks', 'Python Marks']].sum()

# in total(row-wise)
```

```python
# Multiple Column sum() in column-wise

df[['Data Structure Marks','Algorithm Marks','Python Marks']].sum(axis=1)
```

### 2. Correlatoin Between Columns

- Describes the linear relationship between two quantitative variables (numerical Columns).

- Positive correlation:
    - one variable increases, then other variabel also increase.


- Negative correlation:
    - one variable increases, then other variable is decrease.


- Syntax of Correlation:
    - df[['col1', 'col2']].corr()

```python
df.head(3)
```

```python
# Example of correlation of Data Structure Marks and Python Marks	

df[['Data Structure Marks','Python Marks']].corr()

# Positive Correlation
```

```python
# Example of correlation of Algorithm Marks	 and Python Marks	
df[['Algorithm Marks','Python Marks']].corr()

# positive correlation
```

```python
# Error case when try to find correlation between non-numerical and numerical column

df[['StudentID','Python Marks']].corr()

## NOTE:
# Correlation must be numerical column
```

#### Using iloc for Column Range and Row Operations

- Syntax:
    - df['new_col'] = df.iloc[row_idx, col_idx],sum(axis = 1)

    - new_col => add new column in df
    - row_idx star, row_idx_end

    - col_idx start, coln_idx end

```python
df.head(2)
```

```python
df['Total Marks'] = df.iloc[::, 2:5].sum(axis =1)
df.head(3)

# :: -> all rows
# 2 -> start coln idx
# 5 -> end coln idx (not include)
# axis = 1 -> column-wise sum
```

### 3. Aggregate Multiple Statistics at Once

```python
df[["Data Structure Marks", "Algorithm Marks", "Python Marks"]].agg(["sum", "mean", "min", "max", "std"])
```
