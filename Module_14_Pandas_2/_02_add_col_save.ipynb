```python
import pandas as pd
```

```python
df = pd.read_csv("student_data.csv")
df.head(2)
```

### 1. Add a Column with constant value

- Syntax : Add a column with constant value
    - df["New_column_name"] = constant value

```python
df["Country"] = 'Bangladesh'
df.head(3)
```

### 2. Add a Column with some Mathematical Operations

- syntax:
    - df["new_coln_name"] = df["col1"] + df["col2"] + ...

```python
df["Total Marks"] = df['Python Marks'] + df['Data Structure Marks'] + df['Algorithm Marks']

df.head(2)
```

```python
df["Total Marks"].head(8)

# NOTE 
# if one column of (Python Marks,Data Structure Marks, Algorithm Marks) contain NaN value then Total Marks count only NaN value (not sum)
```

### 3. Add a New column with condition (np.where)

```python
# using numpy function [ np.where() ]
import numpy as np

df["A+ in Python"] = np.where(df['Python Marks']>80, 'A+', 'A')

df.head(10)


# if Python Marks is greater than 90 then -> A+ else A
```

### 4. Add a New column with spliting data.

```python
# add a new colun with First name

df['FirstName'] = df['FullName'].str.split(' ').str[0]
df.head(3)
```

```python
# add a new data with last name
df['Last_Name'] = df['FullName'].str.split(' ').str[1]
df.head(2)

# here, split by space
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "salary": [50000, 60000, 55000]
})

# Add a constant column
df1["country"] = "Bangladesh"
print(df1)
```

```python
# new column add with customize value

df1["email"] = ["alice@gmail.com", "bob@yahoo.com", "charlie@outlook.com"]
```

```python
df1 = df1[['username', 'domain']] = df1['email'].str.split('@', expand = True)

df1.head(3)

# str.split("@", expand=True) splits the string into two parts.

# expand=True returns multiple columns.
```

```python

# Split email into username and domain
df1[["username", "domain"]] = df1["email"].str.split("@", expand=True)
print(df1)
```
