### 3. Basic of 2-D plot

- A line plot connects data points with a line that perfect for showing trends over time or continuous data.

- How a data (value) changes over time.

- Example:
    - Temperature data
    - stock market data
    - time series data

- Syntax:
    - plt(x_axis, y_axis)

    - x_axis => always time
    - y_axis => trends

- condition:
    - here, x_axis value and y_axis value must be same length.
    - if only x_axis value given then automatically count the y_axis value (len(x_axis))

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Example of 2-D plot

hours = [2, 3, 4, 1, 5, 7, 2]
days  = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)

plt.show()
```

```python
## labeling x_axis , y_axis and title of the plot

plt.xlabel("Days")
plt.ylabel("Hours")

plt.title("Trend of a Student study time over a Week")

plt.plot(days, hours)

plt.show()
```

#### Multiple lines on One Graph

```python
s1_hours = [3, 6, 4, 10, 8, 4, 3]

s2_hours = [2, 1, 8, 3, 6, 2, 10]

plt.xlabel("Days")

plt.ylabel("Study Hours")

plt.title("Trend of two Students Study time over on a Week")

plt.plot(days, s1_hours)

plt.plot(days, s2_hours)


plt.show()
```

### 4. Customizing Line Plot (color, linestyle, marker and grid etc)

#### Adding the labels using plt.legend()

- Shows labels for multiple lines

- It's identify the each line marker

- for use this we have to use label parameter inside the plt.plot() that stores a name for that line

    - eg: plt.plot(  , label = "Programming")

- plt.legend(loc = 'best')
    - loc => upper left, lower left, lower right

```python
# identity the each line using plt.legend()

plt.xlabel("year")

plt.ylabel("count enrollment")

plt.title("Student enrollment of two course over 8 years")


plt.plot(df['Year'], df['Programming'], label = 'Programming')

plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing')

# plt.legend()- Shows labels for multiple lines

plt.legend()

plt.show()
```

```python
# identity the each line using plt.legend() and customize the location

plt.xlabel("year")

plt.ylabel("count enrollment")

plt.title("Student enrollment of two course over 8 years")


plt.plot(df['Year'], df['Programming'], label = 'Programming')

plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing')

# plt.legend()- Shows labels for multiple lines

plt.legend(loc = 'lower right')

plt.show()
```

#### Customizing line Plot
- inside plt.plot(parameter)

- when parameter colour:
    - 'r', 'g', 'red', 'hex_value'
    - (eg: color='#FF5733')

- when parameter linestyle
    - solid (-) [By default]
    - dashed(---)
    - dashdot(.-.-.-)
    - dotted(.....)
    
    - eg: linestyle = 'dashed'

- when parameter linewidth
    - linewidth = n ( n = 1, 2, 3, 4, ....)
    - eg: linewidth = 3

- when parameter marker
    - 'o', '^', 's', '*', 'D', 'x'
    - eg: marker = 'o'


plt.grid() => adds grid lines


```python
df.head(2)
```

```python
# customize the plt-> parameter or line plot


plt.xlabel("Year")
plt.ylabel("Count of Enrollment")
plt.title('Student enrollment over 8 years')

plt.plot(df['Year'], df['Programming'], label = 'Programming', color = 'b', linestyle='dotted', linewidth= 5,marker = '*')


# plt.grid()
plt.legend()
plt.show()
```

```python
# Customize the 2-D line plot

plt.xlabel("Year")

plt.ylabel("Enrollment Students")

plt.title("Two Course Enrollment Students over 8 Years")

# plot for Programmng course
plt.plot(df['Year'], df['Programming'], label = 'Programming', color='#29055e',linestyle = 'dashed', linewidth = 3,marker = 'o')

# plot for Digital Marketing course
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color = '#073d05', linestyle='dotted',linewidth = 4,marker = '^')

plt.grid()
plt.legend()
plt.show()
```
